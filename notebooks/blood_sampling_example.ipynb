# BloodSampling Example

This is an example of how to use the `BloodSampling` class to parse and use the data form the BloodSampling excel files.

- Import the data parsers

```python
import aifarming.data as d
```

- Load the data from the input files and parse it into a more convenient format. (using the BloodSamplingData class)

```python
file_path = "../sample_files/blood_sampling/only_one/Blood sampling - data 22 weeks.xls"
bloodData = d.BloodSamplingData(file_path)
```

```python
# to acces the data frame use the df attribute
bloodData.df.shape
bloodData.df.sample(10)
```

```python
# I also managed to load in the metadata from the excel file, and it has some useful information
print('company:\n', bloodData.meta.company_name)
print('\nmeasurement date:\n', bloodData.meta.measurement_date)
print('\nweek number:\n', bloodData.meta.week_number)
```
